```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
# samples = np.random.normal(size=10000)
samples = np.random.normal(loc=10,scale=2,size=10000)
sns.distplot(samples)
```

```
s = samples[samples >0] +1
sns.distplot(s)
```

```
new = s ** 3
sns.distplot(new)
```

```
new = s ** (1/5)
sns.distplot(new)
```

```
new = np.power(2,s) # 指數函式
sns.distplot(new)
```
