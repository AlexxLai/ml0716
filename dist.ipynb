```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
# samples = np.random.normal(size=10000)
samples = np.random.normal(loc=10,scale=2,size=10000)
sns.distplot(samples)
```

```
s = samples[samples >0] +1
sns.distplot(s)
```

```
# 找偏度的函式庫
from scipy.stats import skew
# boxcox_normmax(new)
# boxcox_normmax找出lamdan→一般而言→做boxcox轉換
#             →如果lamda趨近於0→取log
new = s ** 5
sns.distplot(new)
print("skew:", skew(new))
```

```
from scipy.stats import boxcox_normmax, boxcox
lamda = boxcox_normmax(new)
print(lamda)
new2 = boxcox(new,lamda)
sns.distplot(new2)
print(skew(new2))
```

```
new = s ** (1/5)
sns.distplot(new)
print("skew:", skew(new))
```

```
lamda = boxcox_normmax(new)
print(lamda)
new2 = boxcox(new, lamda)
sns.distplot(new2)
print(skew(new2))
```

```
new = np.power(2,s) # 指數函式
sns.distplot(new)
print("skew:",skew(new))
```

```
import numpy as np
lamda = boxcox_normmax(new)
print(lamda)
# new2 = boxcox(new, lamda)
new2 = np.log(new)
sns.distplot(new2)
print(skew(new2))
```
